```python
import pandas as pd
import numpy as np
```

```python
students = pd.read_csv('./materiais/StudentsPerformance.csv')
```

```python
provas = students.columns.to_list()[-3:]
```

```python
students['mean'] = students[provas].mean(axis=1).round(2)
```

Obtenção de estatistica descritiva de acordo com variaveis categoricas do dataset

```python
students.groupby(by='gender')['math score'].mean()
```

Supondo que a aprovação se da por notas acima de 70 pontos: 

```python
students.groupby(by='gender')['math score'].agg([min, np.mean, np.std, np.median, max])
```

```python
students.groupby('gender')[provas].agg(np.mean)
```

Agrupamento por genero e lunch (tipo de refeição do aluno)

```python
metrics = [min, np.mean, np.std, np.median, max]
```

```python
students.groupby(by=['gender','lunch'])[provas].agg(metrics)
```

```python
students.groupby('parental level of education')[provas].mean()
```
